**id**:	id\
**study**:	1 for all patients\
**rx**:	Treatment - Obs(ervation), Lev(amisole), Lev(amisole)+5-FU\
**sex**:	1=male\
**age**:	in years\
**obstruct**:	obstruction of colon by tumour\
**perfor**:	perforation of colon\
**adhere**:	adherence to nearby organs\
**nodes**:	number of lymph nodes with detectable cancer\
**time**:	days until event or censoring\
**status**:	censoring status\
**differ**:	differentiation of tumour (1=well, 2=moderate, 3=poor)\
**extent**:	Extent of local spread (1=submucosa, 2=muscle, 3=serosa, 4=contiguous structures)\
**surg**:	time from surgery to registration (0=short, 1=long)\
**node4**:	more than 4 positive lymph nodes\
**etype**:	event type: 1=recurrence, 2=death

```r
colon_data <- read_tsv('/content/colon_dataset.tsv') %>% mutate(etype = recode(etype, `1` = 0, `2` = 1))
colon_data
```

```r
#Разделим датасет по признаку etype.
colon_event_1 <- colon_data %>% filter(etype == 0)
colon_event_2 <- colon_data %>% filter(etype == 1)
```

```r
head(colon_event_1)
```

```r
head(colon_event_2)
```

```r
# Н0-Степень распространенность опухоли не влияет на выживаемость(датасет полностью). р-значение меньше 0,05 - гипотезу отвергаем.

survdiff(Surv(time, status)~extent, data=colon_data)
```

```r
sfit <- survfit(Surv(time, status)~extent, data=colon_data)

survminer::ggsurvplot(sfit, conf.int=TRUE, pval=TRUE, pval.method = TRUE, risk.table=TRUE,
           legend.labs=c("submucosa", "muscle", "serosa", "contiguous structures"), legend.title="Extent",
           palette=c("dodgerblue2", "orchid2", "green", "blue"),
           title="Kaplan-Meier Curve for Colon Cancer Survival",
           risk.table.height=.3)
```

```r
#  Н0-Схема терапии не влияет на выживаемость(датасет e_type==0). р-значение меньше 0,05 - гипотезу отвергаем.

sfit_1 <- survfit(Surv(time, status)~rx, data=colon_event_1)
survdiff(Surv(time, status)~rx, data=colon_event_1)
```

```r
survminer::ggsurvplot(sfit_1, conf.int=TRUE, pval=TRUE, pval.method = TRUE, risk.table=TRUE,
           legend.labs=c("Levamisole", "Levamisole+5-FU", "Observation"), legend.title="Treatment",
           palette=c("dodgerblue2", "orchid2", "yellow"),
           title="Kaplan-Meier Curve for Colon Cancer Survival",
           risk.table.height=.3)
```

```r
fit_2 <- coxph(Surv(time, status)~rx, data=colon_event_1)# комбинированная терапия повышает выживаемость
summary(fit_2)
```

```r
#  Н0-Дифференцированность опухоли не влияет на выживаемость(датасет e_type==1). р-значение меньше 0,05 - гипотезу отвергаем.

sfit_2 <- survfit(Surv(time, status)~differ, data=colon_event_2)
survdiff(Surv(time, status)~differ, data=colon_event_2)
```

```r
survminer::ggsurvplot(sfit_2, conf.int=TRUE, pval=TRUE, pval.method = TRUE, risk.table=TRUE,
           legend.labs=c("well", "moderate", "poor"), legend.title="Differentiation of tumour",
           palette=c("dodgerblue2", "orchid2", "green"),
           title="Kaplan-Meier Curve for Colon Cancer Survival",
           risk.table.height=.3)
```

```r
# Mодель для анализа нескольких переменных (датасет e_type==1). Максимальное влияние на выживаемость оказывает метастазирование в более, чем 4 ЛУ.
fit <- coxph(Surv(time, status)~node4+differ+extent, data=colon_event_2)
summary(fit)
```
